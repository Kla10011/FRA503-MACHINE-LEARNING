read in the dataset

```python
import pandas as pd
df = pd.read_csv('iris.data',header = None )
df.columns = ["1", "2", "3", "4","class"]
```

```python
df.head()
```

```python
import seaborn as sns
sns.pairplot(df,hue="class")
```

training set,teat set

```python
from sklearn.model_selection import train_test_split

training_data, testing_data = train_test_split(df, test_size=0.2, random_state=25)
```

```python
training_data.head() #Random state =25
```

```python
training_data.head() #Random state = 42
```

decision tree

```python
training_data.head() 
```

```python
X = training_data.iloc[:,0:4]
Y = training_data.iloc[:,4]
```

```python
from sklearn import tree
clf= tree.DecisionTreeClassifier()
clf =clf.fit(X,Y)
```

The Result
- Model
- Confusion

```python
X_test = testing_data.iloc[:,0:4]
Y_test = testing_data.iloc[:,4]
```

```python
result = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test, result)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, result))
```

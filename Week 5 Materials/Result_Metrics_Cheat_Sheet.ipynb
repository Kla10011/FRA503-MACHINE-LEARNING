# Result Metrics Cheat Sheet

## Regression

### Mean Squared Error

Full API at: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_squared_error.html

```python
from sklearn.metrics import mean_squared_error


y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]
error = mean_squared_error(y_true, y_pred)

print(error)
```

### Root Mean Squared Error

```python
from sklearn.metrics import mean_squared_error


y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]
error = mean_squared_error(y_true, y_pred, squared = False)

print(error)
```

## Classification

### Confusion Matrix

For simple confusion matrix, you can get full API at https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html

For classification report, you can get full API at https://scikit-learn.org/stable/modules/generated/sklearn.metrics.classification_report.html

```python
from sklearn.metrics import confusion_matrix
y_true = [2, 0, 2, 2, 0, 1]
y_pred = [0, 0, 2, 2, 0, 2]
confusion_matrix(y_true, y_pred)
```

```python
from sklearn.metrics import classification_report
y_true = [2, 0, 2, 2, 0, 1]
y_pred = [0, 0, 2, 2, 0, 2]
target_names = ['class 0', 'class 1', 'class 2']
print(classification_report(y_true, y_pred, target_names=target_names))
```

For ROC curve, you can follow this example: https://scikit-learn.org/stable/auto_examples/model_selection/plot_roc.html#sphx-glr-auto-examples-model-selection-plot-roc-py

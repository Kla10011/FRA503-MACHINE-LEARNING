```python
dff = pd.get_dummies(klaza)
dff.head()
```

```python
from sklearn.model_selection import train_test_split
training_data, testing_data = train_test_split(dff, test_size=0.2, random_state=25)
training_data.head()
```

```python
X = training_data.iloc[:,0:24]
Y = training_data.iloc[:,25]
# X.head()
Y.head()
```

```python
from sklearn import tree
clf= tree.DecisionTreeClassifier()
clf =clf.fit(X,Y)
```

```python
X_test = testing_data.iloc[:,0:24]
Y_test = testing_data.iloc[:,25]
X_test.head()
```

```python
result = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test, result)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, result))
```

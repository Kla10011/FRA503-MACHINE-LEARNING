```python
df_Categorical = df.drop(["age","balance","day","duration","campaign","pdays","previous"],axis=1)
df_Onehot_Cat = pd.get_dummies(df_Categorical)
df_Onehot_Cat_y = df_Onehot_Cat.join([df_y])
df_Onehot_Cat_y.head()
```
